```python
# Pivot: species as columns, using catch numbers
pivot_pa = fish_df.pivot_table(
    index="ID",
    columns="COMNAME",          # or "COMNAME"/"SCINAME" if you want names instead of codes
    values="EXPCATCHNUM",
    aggfunc="sum",
    fill_value=0
)

# Convert to presence (1) / absence (0)
pivot_pa = (pivot_pa > 0).astype(int)

# Reset index if you want ID back as a column
pivot_pa = pivot_pa.reset_index()
```

```python
nefsc_dates_df['ID'] = nefsc_dates_df['ID'].astype(str)
pivot_pa['ID'] = pivot_pa['ID'].astype(str)
```

```python
fish_environment = pd.merge(pivot_pa,nefsc_dates_df,on="ID", how='left').sort_index()
```
