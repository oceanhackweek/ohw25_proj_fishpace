```python
## Removing Columns that are duplicated columns with the ocean data
nefsc_count = nefsc_count.drop(columns=["CRUISE6","CRUISE",'TOW','STATION','STRATUM','STATUS_CODE',"CATCH_COMMENT","CATCHSEX"])
```

```python
## Now you should have 4 columns ID, SVSPP, EXPCATCHNUM, EXPCATCHWT
nefsc_count
```

```python
## Merging the Ocean and count data together into 1 DataFrame
nefsc_full = pd.merge(nefsc_ocean,nefsc_count,on="ID", how='left').sort_index()
```

```python
## There is an error in some dates in the EST_DATES column, so we are creating a column of years based on the first 4 characters of the ID
nefsc_full['year'] = nefsc_full['ID'].astype(str).str[:4].astype(int)
```

```python
# Dropping the years that we don't need. This project focused on 2023 and 2024

drops = np.arange(1963,2023)
df2 = nefsc_full.set_index('year')
nefsc = df2.drop(drops, errors='ignore')
```

```python
# Pivot: species as columns, using catch numbers
pivot_pa = fish_df.pivot_table(
    index="ID",
    columns="COMNAME",          # or "COMNAME"/"SCINAME" if you want names instead of codes
    values="EXPCATCHNUM",
    aggfunc="sum",
    fill_value=0
)

# Convert to presence (1) / absence (0)
pivot_pa = (pivot_pa > 0).astype(int)

# Reset index if you want ID back as a column
pivot_pa = pivot_pa.reset_index()
```

```python
pivot_pa
```

```python
fish_environment = pd.merge(pivot_pa,nefsc_dates_df,on="ID", how='left').sort_index()
```
